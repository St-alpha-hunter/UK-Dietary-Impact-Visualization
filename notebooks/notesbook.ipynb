```python
df = pd.read_csv(get_data_path("result.csv"))
##the average consume between different sources
start = df.columns.get_loc("mean_ghgs")
end = df.columns.get_loc("mean_acid") + 1  # +1 the slice does not include the 
for col in df.columns[start:end]:
    df[col] = df[col].astype("float")
    print(f"{col} average-value: {df[col].mean()}")


start = df.columns.get_loc("mean_ghgs")
end = df.columns.get_loc("mean_acid") + 1  # +1 
for col in df.columns[start:end]:
    df[col] = df[col].astype("float")
    print(f"{col} average-vaule: {df.groupby('diet_group')[col].mean( )}")
```

```python
start = df.columns.get_loc("mean_ghgs")
end = df.columns.get_loc("mean_acid") + 1  # include the final

mean_table = {}

for col in df.columns[start:end]:
    df[col] = df[col].astype(float)
    group_means = df.groupby("diet_group")[col].mean()
    mean_table[col] = group_means

# 转置为行=指标，列=群体
mean_df = pd.DataFrame(mean_table).T

# 可选：重新排序列
mean_df = mean_df[["meat", "meat50", "meat100", "fish", "vegan", "veggie"]]

# 导出为 Excel 文件
mean_df.to_excel("nutrition_means_by_group.xlsx")
```
